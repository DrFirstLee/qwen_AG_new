```python
import os
from tqdm import tqdm

# Í≤ÄÏÉâÏùÑ ÏãúÏûëÌï† ÏµúÏÉÅÏúÑ ÎîîÎ†âÌÜ†Î¶¨
root_dir = "/home/DATA/AGD20K/Seen/trainset/exocentric"
# Í≥†Ïú†Ìïú Ïù¥ÎØ∏ÏßÄÎ•º Ï†ÄÏû•ÌïòÍ∏∞ ÏúÑÌï¥ setÏùÑ ÏÇ¨Ïö©Ìï©ÎãàÎã§.
# (object_name, image_base_name) ÌäúÌîåÏùÑ Ï†ÄÏû•Ìï©ÎãàÎã§.
unique_images_set = set()

print(f"Scanning for unique image base names in: {root_dir}\n")

if not os.path.isdir(root_dir):
    print(f"Error: Directory not found at {root_dir}")
else:
    # os.walk()Îäî root_dirÎ∂ÄÌÑ∞ ÏãúÏûëÌïòÏó¨ Î™®Îì† ÌïòÏúÑ Ìè¥ÎçîÎ•º ÏàúÌöåÌï©ÎãàÎã§.
    # tqdmÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÏßÑÌñâÎ•† ÌëúÏãúÏ§ÑÏùÑ ÌëúÏãúÌï©ÎãàÎã§.
    for dirpath, dirnames, filenames in tqdm(os.walk(root_dir), desc="Scanning"):
        
        # --- action/object Ìè¥ÎçîÏù∏ÏßÄ ÌôïÏù∏ÌïòÎäî Î°úÏßÅ ---
        relative_path = os.path.relpath(dirpath, root_dir)
        parts = relative_path.split(os.sep)
        
        # action/object 2Îã®Í≥Ñ ÍπäÏù¥Ïùò ÎîîÎ†âÌÜ†Î¶¨Ïù∏ÏßÄ ÌôïÏù∏
        if len(parts) == 2:
            action_name, object_name = parts
        else:
            # exocentric/ ÏµúÏÉÅÏúÑ Ìè¥Îçî Îì±ÏùÄ Í±¥ÎÑàÎúÅÎãàÎã§.
            continue
        # --- Î°úÏßÅ ÎÅù ---

        for file in filenames:
            # ÌååÏùº ÌôïÏû•ÏûêÍ∞Ä .jpgÏù∏ÏßÄ ÌôïÏù∏ (ÎåÄÏÜåÎ¨∏Ïûê Íµ¨Î∂Ñ ÏóÜÏùå)
            if file.lower().endswith('.jpg'):
                
                # --- Î≤†Ïù¥Ïä§ Ïù¥Î¶Ñ Ï∂îÏ∂ú Î°úÏßÅ ---
                # ÌååÏùºÎ™ÖÏóêÏÑú action Ï†ëÎëêÏÇ¨(Ïòà: 'carry_')Î•º Ï†úÍ±∞
                prefix = f"{action_name}_"
                if file.startswith(prefix):
                    image_base_name = file[len(prefix):]
                else:
                    image_base_name = file # ÏòàÏô∏ Ï≤òÎ¶¨
                # --- Î°úÏßÅ ÎÅù ---
                
                # (Í∞ùÏ≤¥Î™Ö, Î≤†Ïù¥Ïä§ Ïù¥Î¶Ñ) ÌäúÌîåÏùÑ setÏóê Ï∂îÍ∞Ä
                unique_images_set.add((object_name, image_base_name))

    print("\n--- Scan Complete ---")
    print(f"Total unique image base names found: {len(unique_images_set)}")
```

```python
import os
from collections import defaultdict
from tqdm import tqdm # ÏßÑÌñâ ÏÉÅÌô©ÏùÑ ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú Î≥¥Ïó¨Ï£ºÍ∏∞ ÏúÑÌïú ÎùºÏù¥Î∏åÎü¨Î¶¨

def find_image_cooccurrence(base_directory: str):
    """
    Ï£ºÏñ¥ÏßÑ ÎîîÎ†âÌÜ†Î¶¨ ÌïòÏúÑÏùò action/object Ìè¥ÎçîÎ•º Ïä§Ï∫îÌïòÏó¨
    ÎèôÏùºÌïú Í∞ùÏ≤¥Ïóê ÎåÄÌï¥ Ïó¨Îü¨ actionÏóêÏÑú Ï§ëÎ≥µ ÏÇ¨Ïö©Îêú Ïù¥ÎØ∏ÏßÄÏùò 
    'Î≤†Ïù¥Ïä§ Ïù¥Î¶Ñ'ÏùÑ Ï∞æÎäî ÌååÏù¥Ïç¨ ÏΩîÎìúÏûÖÎãàÎã§.
    
    Ïòà:
    - .../carry/skateboard/carry_skateboard_001.jpg
    - .../hold/skateboard/hold_skateboard_001.jpg
    -> Ïù¥ Îëê ÌååÏùºÏùÄ 'skateboard_001.jpg'ÎùºÎäî Î≤†Ïù¥Ïä§ Ïù¥Î¶ÑÏùÑ Í≥µÏú†ÌïòÎäî Í≤ÉÏúºÎ°ú Í∞ÑÏ£ºÌï©ÎãàÎã§.

    Args:
        base_directory (str): Í≤ÄÏÉâÏùÑ ÏãúÏûëÌï† ÏµúÏÉÅÏúÑ ÎîîÎ†âÌÜ†Î¶¨
                              (Ïòà: /home/DATA/AGD20K/Seen/trainset/exocentric)
    """
    
    print(f"Scanning directory: {base_directory}\n")
    
    # ÌÇ§: (object_name, image_base_name)
    # Í∞í: {action1, action2, ...} (set)
    # Ïòà: [('skateboard', 'skateboard_000089.jpg')] = {'carry', 'hold'}
    image_map = defaultdict(set)
    
    # 1. Î™®Îì† ÌååÏùº Ïä§Ï∫îÌïòÏó¨ Îßµ ÏÉùÏÑ±
    for dirpath, dirnames, filenames in tqdm(os.walk(base_directory), desc="Scanning folders"):
        # dirpath: .../exocentric/carry/skateboard
        # base_directory: .../exocentric
        
        relative_path = os.path.relpath(dirpath, base_directory)
        parts = relative_path.split(os.sep)
        
        # action/object 2Îã®Í≥Ñ ÍπäÏù¥Ïùò ÎîîÎ†âÌÜ†Î¶¨Ïù∏ÏßÄ ÌôïÏù∏
        if len(parts) == 2:
            action_name, object_name = parts
            
            for image_filename in filenames:
                if not image_filename.lower().endswith(('.jpg', '.png', '.jpeg')):
                    continue
                
                # --- ÌïµÏã¨ Î°úÏßÅ ---
                # ÌååÏùºÎ™ÖÏóêÏÑú action Ï†ëÎëêÏÇ¨(Ïòà: 'carry_')Î•º Ï†úÍ±∞ÌïòÏó¨ Î≤†Ïù¥Ïä§ Ïù¥Î¶Ñ ÏÉùÏÑ±
                # Ïòà: "carry_skateboard_000089.jpg" -> "skateboard_000089.jpg"
                prefix = f"{action_name}_"
                if image_filename.startswith(prefix):
                    image_base_name = image_filename[len(prefix):]
                else:
                    # ÏòàÏÉÅÏπò Î™ªÌïú ÌååÏùº Ïù¥Î¶Ñ ÌòïÏãù (Ïòà: actionÍ≥º Ïù¥Î¶ÑÏù¥ Î∂àÏùºÏπò)
                    # Ïù¥ Í≤ΩÏö∞, ÌååÏùºÎ™Ö ÏûêÏ≤¥Î•º Î≤†Ïù¥Ïä§ Ïù¥Î¶ÑÏúºÎ°ú ÏÇ¨Ïö© (ÎπÑÍµêÏóêÏÑú Ï†úÏô∏Îê† Ïàò ÏûàÏùå)
                    image_base_name = image_filename

                # (Í∞ùÏ≤¥Î™Ö, Î≤†Ïù¥Ïä§ ÌååÏùºÎ™Ö)ÏùÑ Í≥†Ïú† ÌÇ§Î°ú ÏÇ¨Ïö©
                image_key = (object_name, image_base_name)
                # Ïù¥ Ïù¥ÎØ∏ÏßÄÍ∞Ä Î∞úÍ≤¨Îêú actionÏùÑ setÏóê Ï∂îÍ∞Ä
                image_map[image_key].add(action_name)

    # 2. Ï§ëÎ≥µÎêú Ïù¥ÎØ∏ÏßÄ(Ïó¨Îü¨ actionÏóê Îì±Ïû•Ìïú Ïù¥ÎØ∏ÏßÄ) ÏßëÍ≥Ñ
    print("\nScanning complete. Analyzing duplicates...")
    
    # ÌÇ§: object_name
    # Í∞í: [(image_base_name, {action1, action2}), ...]
    duplicates_by_object = defaultdict(list)
    
    for (object_name, image_base_name), action_set in image_map.items():
        # 2Í∞ú Ïù¥ÏÉÅÏùò actionÏóêÏÑú Î∞úÍ≤¨Îêú Í≤ΩÏö∞ = Ï§ëÎ≥µ
        if len(action_set) > 1:
            duplicates_by_object[object_name].append(
                (image_base_name, sorted(list(action_set)))
            )

    # 3. Í≤∞Í≥º Ï∂úÎ†•
    print("\n" + "="*50)
    print("--- üì∏ Co-occurrence Statistics Result ---")
    print(" (ÎèôÏùºÌïú Í∞ùÏ≤¥, ÎèôÏùºÌïú Î≤†Ïù¥Ïä§ Ïù¥Î¶ÑÏùÑ Í∞ÄÏ°åÏúºÎÇò Îã§Î•∏ ÌñâÎèô Ìè¥ÎçîÏóê ÏûàÎäî Ïù¥ÎØ∏ÏßÄÎì§)")
    print("="*50)
    
    if not duplicates_by_object:
        print("\nNo image co-occurrences found across different actions.")
        return

    total_duplicated_images = 0
    
    # Í∞ùÏ≤¥ Ïù¥Î¶Ñ ÏàúÏúºÎ°ú Ï†ïÎ†¨ÌïòÏó¨ Ï∂úÎ†•
    for object_name in sorted(duplicates_by_object.keys()):
        duplicates = duplicates_by_object[object_name]
        count = len(duplicates)
        total_duplicated_images += count
        
        print(f"\n### Object: {object_name} (Found {count} co-occurring base images) ###")
        
        # Í∞Å Ï§ëÎ≥µ Ïù¥ÎØ∏ÏßÄÏôÄ Ìï¥Îãπ action Î™©Î°ù Ï∂úÎ†•
        for image_base_name, actions in duplicates:
            print(f"  - Base Name: {image_base_name}")
            print(f"    Found in actions: {actions}")
            
    print("\n" + "="*50)
    print(f"Summary: Found {total_duplicated_images} unique image base names that are shared across multiple action categories.")
    print("="*50)

```
